```python
import shapely
from shapely.strtree import STRtree
from shapely.geometry import Point, box, Polygon, MultiPolygon,MultiPoint
from shapely.ops import unary_union
```

```python
pts = [Point(i,i) for i in range (10)]
```

```python
mpts = MultiPoint(pts)
```

```python
mpts
```

```python
tree = STRtree(pts)
query_geom = Point(2,2).buffer(.99)
```

```python
marec = query_geom.envelope.exterior
query_geom.exterior.union(mpts).union(marec)
```

```python
[o.wkt for o in tree.query(query_geom)]
```
